## Pré-processamento e bases de treinamento/teste

fazendo o tratamento dados com o onehot de colunas para binário (1 e 0)





```python
rh_data = pd.get_dummies(rh_data, columns=['BusinessTravel','Department', 'EducationField', 'MaritalStatus', 'JobRole', ])
```

definindo a normalização entre escala entre o maior e menor

```python
import sklearn.preprocessing
import datetime
import os
import tensorflow as tf
from sklearn.model_selection import train_test_split
 
```

```python
  from sklearn.preprocessing import MinMaxScaler
  scale = sklearn.preprocessing.MinMaxScaler()
  X = scale.fit_transform(rh_data.drop(['Attrition'], axis=1))
  Y = rh_data['Attrition']

  X_train,X_test, Y_train, Y_test = train_test_split( X, Y, test_size=0.3)
```

# Ultilizando a rede neural artificial / Criando modelo

Modelo MLP com 44 entradas:


*   44 neurônios
*   camada de saida binária para prever a saída da empresa (sim ou não) sigmod

*   duas camadas dense de entrada uma softmax e outra relu
*   Optimização sendo a adam para tratamento binário





```python
  model = tf.keras.models.Sequential()
  model.add(tf.keras.layers.Dense(64, activation='relu', input_dim=48))
  model.add(tf.keras.layers.Dense(32, activation='softmax'))
  model.add(tf.keras.layers.Dense(1, activation='sigmoid'))

  model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
  model.fit(X_train, Y_train, epochs=100, batch_size=32)
```

Testando a acuracia (cerca de 84.8%) e no final com taxa de acertividade de 95%

```python
model.evaluate(X_test, Y_test)
```
